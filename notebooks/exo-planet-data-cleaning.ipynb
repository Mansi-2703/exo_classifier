```python
df2 = pd.read_csv('/content/TESS.csv')
```

```python
display(df['koi_disposition'].value_counts())
```

```python
df2['tfopwg_disp'] = df2['tfopwg_disp'].replace(['CP', 'KP', 'PC', 'APC'], 1)
df2['tfopwg_disp'] = df2['tfopwg_disp'].replace(['FP', 'FA'], 0)
display(df2['tfopwg_disp'].value_counts())
```

```python
df_kepler = df[['koi_disposition', 'koi_period', 'koi_duration', 'koi_depth', 'koi_prad', 'koi_insol', 'koi_teq', 'koi_steff', 'koi_slogg', 'koi_srad', 'ra', 'dec']]
df_tess = df2[['tfopwg_disp', 'pl_orbper', 'pl_trandurh', 'pl_trandep', 'pl_rade', 'pl_insol', 'pl_eqt', 'st_teff', 'st_logg', 'st_rad', 'ra', 'dec']]
```

```python
column_mapping = {
    'tfopwg_disp': 'koi_disposition',
    'pl_orbper': 'koi_period',
    'pl_trandurh': 'koi_duration',
    'pl_trandep': 'koi_depth',
    'pl_rade': 'koi_prad',
    'pl_insol': 'koi_insol',
    'pl_eqt': 'koi_teq',
    'st_teff': 'koi_steff',
    'st_logg': 'koi_slogg',
    'st_rad': 'koi_srad',
    'ra': 'ra',
    'dec': 'dec'
}
df_tess.rename(columns=column_mapping, inplace=True)
display(df_tess.head())
```

```python
combined_df = pd.concat([df_kepler, df_tess], ignore_index=True)
display(combined_df.head())
display(combined_df.shape)
```

```python
# Replace this dictionary with your desired column name mapping
new_column_names = {
    'koi_disposition': 'disposition',
    'koi_period': 'orbital_period_days',
    'koi_duration': 'transit_duration_hours',
    'koi_depth': 'transit_depth_ppm',
    'koi_prad': 'planetary_radius_earth_radii',
    'koi_insol': 'insolation_flux_earth_flux',
    'koi_teq': 'equilibrium_temperature_k',
    'koi_steff': 'stellar_effective_temperature_k',
    'koi_slogg': 'stellar_surface_gravity_log10_cm_s2',
    'koi_srad': 'stellar_radius_solar_radii',
    'ra': 'right_ascension_decimal_degrees',
    'dec': 'declination_decimal_degrees'
}

combined_df.rename(columns=new_column_names, inplace=True)
display(combined_df.head())
```
